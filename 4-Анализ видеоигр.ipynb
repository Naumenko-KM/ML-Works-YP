### 2. Подготовим данные

В колонке `user_score` кроме пропусков есть значение 'tbd' - to be determined, это значит, что для игры не набралось минимальное количество оценок пользователей. Заменим эти значения на NaN и создадим новый столбец `tbd` со значениями True или False. Пропущенные значения заменим на '-1'.

```python
df['tbd'] = df['user_score'] == 'tbd'
df['tbd'].value_counts()
```

```python
df['rating'].fillna('no rate', inplace=True)
df['year_of_release'].fillna(-1, inplace=True)
df['critic_score'].fillna(-1, inplace=True)
```

Приведем данные к нужным типам.

```python
df[['platform', 'genre', 'rating']] = df[['platform', 'genre', 'rating']].astype('category')
df[['year_of_release', 'critic_score']] = df[['year_of_release', 'critic_score']].astype('int')
df['user_score'] = df['user_score'].astype('float')
```

### 5. Проверим гипотезы

##### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Сформулируем нулевую гипотезу $Н_0$: средний рейтинг игр для платформы Xbox One равен среднему рейтингу игр для платформы PC.

Сформулируем альтернативную гипотезу $Н_0$: средний рейтинг игр для платформы Xbox One отличен от  среднего рейтинга игр для платформы PC.

```python
sample1 = df.loc[(df['platform']=='XOne') & (df['user_score']!=-1), 'user_score']
sample2 = df.loc[(df['platform']=='PC') & (df['user_score']!=-1), 'user_score']
```

```python
# Xbox One
sample1.hist(bins=15);
plt.title('Xbox One')
plt.xlabel('user_score')
plt.ylabel('count');
```

```python
# PC
sample2.hist(bins=15);
plt.title('PC')
plt.xlabel('user_score')
plt.ylabel('count');
```

```python
alpha= 0.05

# Добавил тест Левина
levene = st.levene(sample1, sample2)
print(f'Levene\'s p-value: {levene.pvalue:.4%}')

if levene.pvalue < alpha:    
    print('Отвергаем нулевую гипотезу. Считаем, что дисперсии выборок не равны.')
else:
    print('Не получилось отвергнуть нулевую гипотезу. Считаем, что дисперсии выборок равны.')
```

```python
result = st.ttest_ind(sample1, sample2, equal_var=True)

print(f'Xbox One mean user scrore: {sample1.mean():.2}')
print(f'PC mean user scrore: {sample2.mean():.2}')
print(levene.pvalue)
print(f'pvalue: {result.pvalue:.4%}')

if (result.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

На PC пользователи оценивают игры лучше чем на Xbox, возможно, это связано с тем, что некоторые игры присутствуют только на одной из этих двух платформ. 

##### Средние пользовательские рейтинги жанров Action и Sports разные.

Сформулируем нулевую гипотезу $Н_0$: средний рейтинг игр жанра 'Action' равен среднему рейтингу игр жанра 'Sports'.

Сформулируем альтернативную гипотезу $Н_1$: средний рейтинг игр жанра 'Action' отличается от  среднего рейтинга игр жанра 'Sports'.

```python
sample1 = df.loc[(df['genre']=='Action') & (df['user_score']!=-1), 'user_score']
sample2 = df.loc[(df['genre']=='Sports') & (df['user_score']!=-1), 'user_score']
```

```python
sample1.hist(bins=15);
plt.title('Action')
plt.xlabel('user_score')
plt.ylabel('count');
```

```python
sample2.hist(bins=15);
plt.title('Action')
plt.xlabel('user_score')
plt.ylabel('count');
```

```python
alpha= 0.05

levene = st.levene(sample1, sample2)
print(f'Levene\'s p-value: {levene.pvalue:.4%}')

if levene.pvalue < alpha:    
    print('Отвергаем нулевую гипотезую. Считаем, что дисперсии выборок не равны.')
else:
    print('Не получилось отвергнуть нулевую гипотезу. Считаем, что дисперсии выборок равны.')
```

```python
result = st.ttest_ind(sample1, sample2, equal_var=False)

print(f'Action mean user scrore: {sample1.mean():.2}')
print(f'Sports mean user scrore: {sample2.mean():.2}')
print(f'pvalue: {result.pvalue:.4%}')

if (result.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Экшн и спорт - два самых популярных жанра, неудивительно, что средние рейтинги пользователей близки.
